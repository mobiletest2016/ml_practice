```
from sklearn.datasets import load_boston
```

```
data = load_boston()
```

```
import pandas as pd
```

```
boston = pd.DataFrame(data.data, columns=data.feature_names)
```

```
boston['PRICE'] = data.target
```

```
boston
```

```
import seaborn as sns
sns.pairplot(boston)
```

```
X = boston[boston.columns[:-1]]
Y = boston[['PRICE']]
```

```
X
```

```
Y
```

```
from sklearn.model_selection import train_test_split
```

```
trainX, testX, trainY, testY = train_test_split(X, Y)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
```

```
linreg = LinearRegression().fit(trainX, trainY)
linreg.score(testX, testY)
```

```
reg = Ridge().fit(trainX, trainY)
reg.score(testX, testY)
```

```
reg = Lasso().fit(trainX, trainY)
reg.score(testX, testY)
```

```
reg = ElasticNet().fit(trainX, trainY)
reg.score(testX, testY)
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```
reg = RandomForestRegressor().fit(trainX, trainY)
reg.score(testX, testY)
```

```
reg = GradientBoostingRegressor().fit(trainX, trainY)
reg.score(testX, testY)
```

```
from sklearn.neural_network import MLPRegressor
reg = MLPRegressor(alpha=100).fit(trainX, trainY)
reg.score(testX, testY) 
```

```
from sklearn.neural_network import MLPRegressor
reg = MLPRegressor(alpha=20).fit(trainX, trainY)
reg.score(testX, testY) 
```

```
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(MLPRegressor(), param_grid={'alpha': [0.001, 0.01, 0.1, 1, 10, 20, 50, 100]})
clf.fit(trainX, trainY)
```

```
clf.cv_results_
```
